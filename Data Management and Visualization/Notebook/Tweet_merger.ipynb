```python
tweets_completi.head()
```

```python
def correct_score(string):
    m = re.match("([0-9]*) (-) ([0-9]*)", string)
    return m.group(3) + " - " + m.group(1)
```

```python
tweets_completi.loc[tweets_completi["gara"] == 1, "score"]  = tweets_completi[tweets_completi["gara"] == 1].apply\
(lambda row: correct_score(row["score"]), axis = 1)

tweets_completi.loc[tweets_completi["gara"] == 2, "score"]  = tweets_completi[tweets_completi["gara"] == 2].apply\
(lambda row: correct_score(row["score"]), axis = 1)

tweets_completi.loc[tweets_completi["gara"] == 5, "score"]  = tweets_completi[tweets_completi["gara"] == 5].apply\
(lambda row: correct_score(row["score"]), axis = 1)
```
