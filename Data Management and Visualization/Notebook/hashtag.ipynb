```python
import pandas as pd
from itertools import chain
import re
from tqdm import tqdm
```

## GSW

```python
dati = pd.read_csv("C:/Users/fedem/Desktop/Università/DATA_SCIENCE/I_ANNO/DATA_MANAGEMENT_AND_DATA_VISUALIZATION/Progetto/tweet_gsw.csv", engine = "python", encoding="utf-8")
```

```python
def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))

# Calcolo lunghezza split
lens = dati["hashtag"].str.split(",").map(len)
```

```python
hashtags = pd.DataFrame({"boh" : chainer(dati["hashtag"])})
```

```python
hashtags["boh"] = hashtags["boh"].str.replace("[", "")
hashtags["boh"] = hashtags["boh"].str.replace("]", "")
hashtags["boh"] = hashtags["boh"].str.replace("'", "")
hashtags["boh"] = hashtags["boh"].str.replace(" ", "")
```

```python
hashtags["boh"].value_counts()
```

```python
dic = {"hashtag" : hashtags["boh"].value_counts().index, "count_gsw" : hashtags["boh"].value_counts()}
```

```python
gsw_hashtag = pd.DataFrame(dic)
gsw_hashtag.index = range(0, len(gsw_hashtag))
```

## Neutro

```python
dati = pd.read_csv("C:/Users/fedem/Desktop/Università/DATA_SCIENCE/I_ANNO/DATA_MANAGEMENT_AND_DATA_VISUALIZATION/Progetto/tweet_neutro.csv", engine = "python", encoding = "utf-8")
```

```python
hashtags = pd.DataFrame({"boh" : chainer(dati["hashtag"])})
```

## Merge

```python
tot = gsw_hashtag.merge(tor_hashtag, on="hashtag", how="outer")
tot = tot.merge(neutro_hashtag, on="hashtag", how="outer")
```

```python
tot = tot.fillna(0)
```

```python
tot["totale"] = tot["count_gsw"] + tot["count_neutro"] + tot["count_tor"]
```
