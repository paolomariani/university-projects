```python
tweets_completi["id"] = tweets_completi["id"].astype("int64")
tweets_completi["retweet_id"] = tweets_completi["retweet_id"].fillna(-1).astype("int64")
```

```python
retweet_count = pd.DataFrame(data = [tweets_completi.groupby("retweet_id")["id"].count().sort_values(ascending=False).index, tweets_completi.groupby("retweet_id")["id"].count().sort_values(ascending=False).values])
```

```python
retweet_count = retweet_count.T
retweet_count.columns = ["retweet_id", "retweet_count"]
```

```python
retweet_count.head()
```

```python
retweet_count = retweet_count.drop(0).reset_index(drop = True)
```

```python
retweet_count.head()
```

```python
tweets_completi = tweets_completi.join(retweet_count.set_index("retweet_id"), on = "retweet_id")
```

```python
tweets_completi.sort_values("retweet_count", ascending=False)
```

```python
tweets_completi["retweet_count"] = tweets_completi["retweet_count"].fillna(0)
tweets_completi["retweet_count"] = tweets_completi["retweet_count"].replace(518658, 0)
```

```python
tweets_completi.sort_values("retweet_count", ascending=False)
```

```python
retweet = tweets_completi[["id", "text", "retweet_id", "retweet_count", "date", "time", "device", "sentiment", "team", "gara", "quarter"]]
```

```python
retweet["date_time"] = retweet["date"].map(str) + retweet["time"]
```

```python
prova = retweet.sort_values("date_time").drop_duplicates(subset= "retweet_id", keep = "first")
```

```python
prova = prova.drop(0)
prova = prova.reset_index(drop=True)
```

```python
prova.sort_values("retweet_count", ascending=False)
```
